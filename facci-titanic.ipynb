```python
df = df.rename(columns={'2urvived': 'Survived'})
```

```python
df.groupby('Sex')['Survived'].mean()
```

```python
df.groupby('Pclass')['Survived'].mean()
```

```python
import matplotlib.pyplot as plt

df.groupby('Sex')['Survived'].mean().plot(kind='bar')
plt.title('Taxa de Sobrevivência por Sexo')
plt.ylabel('Taxa de Sobrevivência')
plt.show()
```

```python
df['AgeGroup'] = pd.cut(df['Age'], bins=[0,12,18,35,60,100],
                        labels=['Criança','Adolescente','Jovem Adulto','Adulto','Idoso'])

df.groupby('Age')['Survived'].mean()
```

## Análise de Sobrevivência por Sexo

A análise mostra que a taxa média de sobrevivência das mulheres foi significativamente maior que a dos homens.

Isso indica que, durante o desastre do Titanic, houve prioridade no resgate de mulheres, reforçando a política de "mulheres e crianças primeiro".

Essa diferença demonstra como variáveis categóricas podem influenciar fortemente um desfecho.
